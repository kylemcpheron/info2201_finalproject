```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#reading files
#df = pd.read_csv
violation19 = pd.read_csv("data/Housing_Maintenance_Code_Violations_2019.csv")
violation21 = pd.read_csv("data/Housing_Maintenance_Code_Violations_2021.csv")
violation24 = pd.read_csv("data/Housing_Maintenance_Code_Violations_2024.csv")
```

```python
violation19.head()
```

```python
violation21.head()
```

```python
# Combine both years for borough comparison
violation19['Year'] = 2019
violation21['Year'] = 2021
violation24['Year'] = 2024
combined_df = pd.concat([violation19, violation21, violation24])

# Group by Year and Borough
by_year_borough = combined_df.groupby(['Year', 'Borough']).size().unstack()

# Plot
by_year_borough.T.plot(kind='bar', figsize=(8,6), color=['#9B7EDC', '#580F41', '#C4B7D3'])
plt.title("Housing Violations by Borough: 2019 vs 2021 vs 2024")
plt.xlabel("Borough")
plt.ylabel("Violations")
plt.xticks(rotation=45)
plt.legend(title="Year")
plt.tight_layout()
plt.savefig("violations_by_borough_comparison.png")
plt.show()
```
