```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load your cleaned CSVs
df_2019 = pd.read_csv("data/Housing_Maintenance_Code_Violations_2019.csv")
df_2021 = pd.read_csv("data/Housing_Maintenance_Code_Violations_2021.csv")
df_2024 = pd.read_csv("data/Housing_Maintenance_Code_Violations_2024.csv")

# Step 2: Count classes for each year
classes_2019 = df_2019['Class'].value_counts()
classes_2021 = df_2021['Class'].value_counts()
classes_2024 = df_2024['Class'].value_counts()

# Step 3: Combine into a DataFrame
combined = pd.DataFrame({
    '2019': classes_2019,
    '2021': classes_2021,
    '2024': classes_2024
}).fillna(0).astype(int)

# Step 4: Plot
combined.plot(kind='bar', figsize=(8,6), color=['#9B7EDC', '#580F41', '#C4B7D3'])
plt.title("Violation Class Distribution by Year")
plt.xlabel("Violation Class")
plt.ylabel("Number of Violations (out of 1,000)")
plt.xticks(rotation=0)
plt.legend(title="Year")
plt.tight_layout()
plt.savefig("violation_classes_by_year.png")
plt.show()
```
